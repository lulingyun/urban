```python
#关于文档的一个数据分类
import pandas as pd
import seaborn as sns  #用于绘制热图的工具包
from scipy.cluster import hierarchy  #用于进行层次聚类，话层次聚类图的工具包
from scipy import cluster   
import matplotlib.pyplot as plt
from sklearn import decomposition as skldec #用于主成分分析降维的包
import numpy as np
#import csv
import math
from pandas import Series,DataFrame
from matplotlib.ticker import  MultipleLocator
```

```python
####2.计算每一个城市的zipf plot
### 读取1991-2017年投资的数据
# Data=pd.read_excel('E:/文档系列/python 系列/TheData/total/11clean_area_invesments(91-17-RMB).xlsx')
# city=Data['city'].drop_duplicates().values
Data = pd.read_excel('E:/文档系列/python 系列/TheData/total/11clean_area_invesments(91-17-RMB).xlsx')
city = Data['city'].drop_duplicates().values
data2 = pd.read_excel('E:/文档系列/python 系列/TheData/0_theDATA_end(1991-2017pop&GDP&vc)/ALL_Intersection.xlsx')
city2 = data2['City'].values###得到城市
```

```python
cityUN=list(set(city2).intersection(set(city))) ###得到两个城市的交集
```

```python
len(cityUN)
```

```python
################# 三级产业的zipf 分布
dsj=Data.groupby(['city',"三级产业名称"], as_index=False)['投资金额(RMB)'].count()
names=Data['三级产业名称'].drop_duplicates().values###获取一级产业名称
names=np.append(names[0:10],names[11:])
###按照城市得到分类
DATA=[]###按照城市对数据进行分组
for li in cityUN:
    dat=dsj[dsj['city'].isin([li])]
    DATA.append(dat)
###得到城市的序列 {产业，1/0}
INDall=[]###得到所有城市的产业序列
for lend in range(len(DATA)):
    col=DATA[lend].shape[0]###得到行数
    indu=[]###得到一级产业的数量
    for ind in names:
        #yinum=0 ###得到产业的数量
        find=0 ###标值是否存在该产业
        for co in range(col):            
            if DATA[lend].iloc[co:co+1,:]['三级产业名称'].values==ind:
                indu.append(DATA[lend].iloc[co:co+1,:]['投资金额(RMB)'].values[0])
                find=1
        if find==0:
            indu.append(0)
    INDall.append(indu)
dfgs=pd.DataFrame(INDall)###将list转化为dataframe

```

```python
###为了让城市和颜色，拼音对应
cityPin = ['suzhou','hangzhou','chengdu','tianjin','nanjing','shanghai','beijing','wuhan','wuxi','shenzhen','guangzhou']
cityCN = ['苏州','杭州','成都','天津','南京','上海','北京','武汉','无锡','深圳','广州']
c=len(cityPin)###一级产业数量
a=list(sns.color_palette("hls", c))
# yj_color=dict(zip(yjName,a))
PINcolor = dict(zip(cityCN,a))
dicPIN = {'苏州':'suzhou','杭州':'hangzhou','成都':'chengdu','天津':'tianjin','南京':'nanjing','上海':'shanghai','北京':'beijing','武汉':'wuhan','广州':'guangzhou','无锡':'wuxi','深圳':'shenzhen'}
###得到投资多余35的城市的拼音
```

```python
###y轴除以均值试一试
#pp=np.log10(pcnumber)###zipf中取双log
###得到每一个城市的三级产业投资次数zipf分布
###按照投资次数进行排序
plt.rcParams['font.sans-serif']=['Simhei']     #解决中文显示问题，目前只知道黑体可行
plt.rcParams['axes.unicode_minus']=False       #解决负数坐标显示问题

dfgs2=dfgs.fillna(0)
x=np.arange(1,len(names)+1)
fig=plt.figure(figsize=(7,6))
ax=fig.add_subplot(111)
for tick in ax.yaxis.get_major_ticks():
        tick.label1.set_fontproperties('stixgeneral')
plt.tick_params(labelsize=16) #刻度字体大小13
font2 = {'family' : 'Arial',
    'weight' : 'normal',
    'size' : 22,
    }###设置横纵坐标的字体格式
font1 = {'family' : 'Arial',
    'weight' : 'normal',
    'size' : 14,
    }###设置横纵坐标的字体格式
plt.xlabel('rank',font2)
plt.ylabel('P',font2)
cl = 0###记录大于35的城市的个数，方便读取拼音
x35 = []####记录大于35的城市的值
y35 = []
city35 = []
len35 = []###记录城市还含有的三级产业种类数
for c in range(dfgs2.shape[0]):
    pp=np.log10(dfgs2.iloc[c:c+1,:].values[0])
    pp2=[vv for vv in pp if vv>0]###所有大于0的数
    if len(pp2)>0:
        #avg=sum(pp2)/len(pp2)###所有大于0的数的均值
        avg=sum(pp2)###所有大于0的数的均值
        pp3=pp/avg###利用均值处理数据
    else:
        pp3=pp/1
    cp=dict(zip(names,pp3))
    cpSortList = sorted( cp.items(),key = lambda x:x[1],reverse = True)
    y=[]
    X=[]###保存城市投资产业种类数
    for i in range(len(cpSortList)):
        if cpSortList[i][1]>0:
            y.append(cpSortList[i][1])
            X.append(x[i])
    
    
    if len(X) > 35:
        len35.append(len(X))
        x35.append(X)
        y35.append(y)
        city35.append(cityUN[c])
#         plt.plot(X,y)
        cl += 1
    else:
        plt.plot(X[:-1],y[:-1],color='gray',alpha=0.25)
###将三级产业数量大于35的排序
cp35 = dict(zip(city35,len35))
cpSortList35 = sorted( cp35.items(),key = lambda x:x[1],reverse = True)
for mm in range(len(cpSortList35)):
    for nn in range(len(x35)):
        if city35[nn]==cpSortList35[mm][0]:
            plt.plot(x35[nn],y35[nn],color=PINcolor[city35[nn]],label=dicPIN[city35[nn]])#,label=dicPIN[city35[nn]]
#             plt.text(60, 0.11+mm//,dicPIN[city35[nn]] ,size=16)
plt.yscale('log')
plt.legend(prop=font1)
plt.savefig('E:/文档系列/python 系列/picScalling/20201025NEW/Fig1d.pdf', bbox_inches='tight')
```

```python
############ 看大城市zipf的拟合情况 ###########3
from scipy.optimize import leastsq  ##引入最小二乘法算法
from sklearn.metrics import r2_score
#需要拟合的函数func :指定函数的形状
def func(p,x):
    k,b=p
    return k*x+b

##偏差函数：x,y都是列表:这里的x,y更上面的Xi,Yi中是一一对应的
def error(p,x,y):
    return func(p,x)-y
def get_nihe(dfgs2,cityUN,cname):
    xrank=np.arange(1,len(names)+1)
    for c in range(len(cityUN)):
        if cityUN[c]==cname:
            yt = np.log10(dfgs2.iloc[c:c+1,:].values[0])###得到城市的 #industry
            pp2=[vv for vv in yt if vv>0]###所有大于0的数
            if len(pp2)>0:
                pp3 = yt
                #avg=sum(pp2)/len(pp2)###所有大于0的数的均值
#                 avg=sum(pp2)###所有大于0的数的均值
#                 pp3=yt/avg###利用均值处理数据
            else:
                pp3=yt/1
            cp=dict(zip(names,pp3))
            cpSortList = sorted( cp.items(),key = lambda x:x[1],reverse = True)
            Y=[]
            X=[]###保存城市投资产业种类数
            for i in range(len(cpSortList)):
                if cpSortList[i][1]>0:
                    Y.append(cpSortList[i][1])
                    X.append(np.log10(xrank[i]))
    ###最小二乘拟合 ###
    yt1=[]
    yt2=[]
    for i in range(len(Y)):
        if Y[i]==-float('Inf') or X[i]==float('Inf'):
            continue
        else:
            yt1.append(X[i])
            yt2.append(Y[i])
    #k,b的初始值，可以任意设定,经过几次试验，发现p0的值会影响cost的值：Para[1]
    p0=[1,10]
    Xi=np.array(yt1)
    Yi=np.array(yt2)
    #把error函数中除了p0以外的参数打包到args中(使用要求)
    Para=leastsq(error,p0,args=(Xi,Yi))## 三个参数：误差函数、函数参数列表、数据点
    #读取结果
    k,b=Para[0] 
    y = [k*xx+b for xx in yt1]
    k1=round(k, 2)
    r=r2_score(yt2,y)
    r2=round(r, 2)
    return(k1,yt1,y,yt2)   
        
```

```python
plt.figure(figsize=(7,6))
font2 = {'family' : 'Arial',
    'weight' : 'normal',
    'size' : 22,
    }###设置横纵坐标的字体格式
font1 = {'family' : 'Arial',
    'weight' : 'normal',
    'size' : 16,
    }###设置横纵坐标的字体格式
    #设定X,Y轴标签和title
plt.tick_params(labelsize=16) #刻度字体大小13
plt.xlabel('$\log_{10}$'+' (rank)',font2)
plt.ylabel('$\log_{10}$'+' (# industry)',font2)
Cvalue = get_nihe(dfgs2,cityUN,'北京')###得到城市的三级产业排名，数量和拟合值
plt.scatter(Cvalue[1],Cvalue[3])
plt.plot(Cvalue[1],Cvalue[2],label='slope_BJ = {}'.format("%.2f" % Cvalue[0]))
# plt.text(1.5,4.0,'slope_BJ = {}'.format("%.2f" % Cvalue[0]),size=16)

CvalueSH = get_nihe(dfgs2,cityUN,'上海')###得到城市的三级产业排名，数量和拟合值
plt.scatter(CvalueSH[1],CvalueSH[3])
plt.plot(CvalueSH[1],CvalueSH[2],label='slope_SH = {}'.format("%.2f" % CvalueSH[0]))
# plt.text(1.5,3.7,'slope_BJ = {}'.format("%.2f" % CvalueSH[0]),size=16)

CvalueSZ = get_nihe(dfgs2,cityUN,'深圳')###得到城市的三级产业排名，数量和拟合值
plt.scatter(CvalueSZ[1],CvalueSZ[3])
plt.plot(CvalueSZ[1],CvalueSZ[2],label='slope_SZ = {}'.format("%.2f" % CvalueSZ[0]))
# plt.text(1.5,3.4,'slope_BJ = {}'.format("%.2f" % CvalueSZ[0]),size=16)

CvalueGZ = get_nihe(dfgs2,cityUN,'广州')###得到城市的三级产业排名，数量和拟合值
plt.scatter(CvalueGZ[1],CvalueGZ[3])
plt.plot(CvalueGZ[1],CvalueGZ[2],label='slope_GZ = {}'.format("%.2f" % CvalueGZ[0]))

CvalueHZ = get_nihe(dfgs2,cityUN,'杭州')###得到城市的三级产业排名，数量和拟合值
plt.scatter(CvalueHZ[1],CvalueHZ[3])
plt.plot(CvalueHZ[1],CvalueHZ[2],label='slope_HZ = {}'.format("%.2f" % CvalueHZ[0]))
plt.legend(prop=font1)
# plt.yscale('log')
```
