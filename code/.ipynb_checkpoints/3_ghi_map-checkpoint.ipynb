```python
####,moneyClass
cmap = plt.cm.YlOrRd
Data =pd.read_excel('E://python /TheData/0_theDATA_end(1991-2017pop&GDP&vc)/ALL_Intersection(1991-2017).xlsx')
```

```python
dataCij = pd.read_excel('E://python /TheData/0_theDATA_end(1991-2017pop&GDP&vc)/CijClass(2000-2017).xlsx')
city = Data['City'].values
City = dataCij['city'].values
```

```python
dataPOP=Data.iloc[:,2:20]###POP
dataGDP=Data.iloc[:,21:39]###GDP
dataAmount = Data.iloc[:,49:67]###
dataTimes = Data.iloc[:,76:94]###
sumP=dataPOP.sum(axis=1).values
sumG=dataGDP.iloc[:,0:17].sum(axis=1).values
sumA=dataAmount.sum(axis=1).values
sumT=dataTimes.sum(axis=1).values
```

```python
dataTimes[0:1]
```

```python
#### 127GDP\amount
C_POP = []###gddp
C_amount = []
C_GDP = []
C_times = []
for c in City:
    have = 0###
    for i in range(len(city)):
        if city[i]== c:
            C_POP.append(sumP[i]/2)
            C_amount.append(sumA[i]/(68*2))
            C_GDP.append(sumG[i]/(850000*2))
            C_times.append(sumT[i]/2)
            have=1
    if have == 0:
        C_POP.append(0)
        C_amount.append(0)
        C_GDP.append(0)
        C_times.append(0)
```

```python
### 
dataCij['sumPOP'] = C_POP
dataCij['sumAmount'] = C_amount
dataCij['sumGDP'] = C_GDP
dataCij['sumTimes'] = C_times
```

```python
dataCij_A = dataCij.sort_values(by="sumAmount" , inplace=False, ascending=False)
dataCij_T = dataCij.sort_values(by="sumTimes" , inplace=False, ascending=False)
dataCij_G = dataCij.sort_values(by="sumGDP" , inplace=False, ascending=False)
```

```python
AvgP=dataPOP.mean(axis=1).values
AvgG=dataGDP.iloc[:,0:17].mean(axis=1).values
AvgA=dataAmount.mean(axis=1).values
AvgT=dataTimes.mean(axis=1).values
##  avg
GYP = AvgP/AvgP.mean()
GYG = AvgG/AvgG.mean()
GYA = AvgA/AvgA.mean()
GYT = AvgT/AvgT.mean()
# ### sum
# GYP = AvgP/AvgP.sum()
# GYG = AvgG/AvgG.sum()
# GYA = AvgA/AvgA.sum()
# GYT = AvgT/AvgT.sum()
```

```python
### 
#### 127GDP\amount
C_avgPOP = []###gddp
C_avgamount = []### avgAmount/avgGdp
C_avgGDP = []
C_avgtimes = []
for c in City:
    have = 0###
    for i in range(len(city)):
        if city[i]== c:
#             C_avgPOP.append(GYP[i]/AvgP)
#             C_avgamount.append(GYA[i]*150/GYG[i])###### 150,
#             C_avgGDP.append(GYA[i]/AvgA)
#             C_avgtimes.append(GYT[i]*150/GYG[i])
            C_avgamount.append(GYA[i]/GYG[i])####
            C_avgtimes.append(GYT[i]/GYG[i])
            C_avgGDP.append(GYG[i])
            have=1
    if have == 0:
#         C_avgPOP.append(0)
        C_avgamount.append(0)
#         C_avgGDP.append(0)
        C_avgtimes.append(0)
        C_avgGDP.append(0)

# dataCij['sumPOP'] = C_POP
dataCij['avgAmount'] = C_avgamount
# dataCij['sumAmount3'] = C_avgamount
# dataCij['sumGDP'] = C_GDP
dataCij['avgTimes'] = C_avgtimes
dataCij['avgGMP'] = C_avgGDP
# dataCij['sumTimes3'] = C_avgtimes
```
